## 2.Get url and pass to BeautifulSoup

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_districts_of_Bangkok') #ตั้งตัวแปร url แล้วใช้ requests.get()เข้าหน้าเวป
soup = BeautifulSoup(url.text, 'lxml') #ใช้ BeautifulSoupโหลดHTMLจากเวป
```

Preview data in table

```python
data = soup.find_all('table',class_='wikitable sortable')
# for find_all ref: https://www.crummy.com/software/BeautifulSoup/bs4/doc/#searching-by-css-class
```

```python
data
```

```python
th_tag = soup.tbody.find_all('th')

title = []

for item in th_tag:
    title.append(item.get_text().replace('\n',''))

title
```

# Find District Name list 

```python
xbody = soup.tbody.find_all('a')
```

Preview first 10 rows

```python
xbody[:10]
```

# Create list to store District Name

```python
#Create List named district_list
district_list = []
#Extract text from xbody and store in district_list
for i in range(0,len(xbody)):
   district_list.append(xbody[i].get_text())
```

```python
district_list[0:10]
```

```python
post = soup.find_all('td')
```

```python
postal = []

for posts in range(1,len(post),7):
    postal.append(post[posts].get_text().replace('\n',''))
```

```python
descr = soup.find_all('td')

desc_th = []

for des in range(2,len(descr),7):
    desc_th.append(descr[des].get_text().replace('\n',''))
```

```python
desc_th.pop(50)
```

```python
desc_th
```

```python
pop = soup.find_all('td')

pop_th = []

for p in range(3,len(pop),7):
    pop_th.append(pop[p].get_text().replace('\n',''))
```

```python
pop_th
```

```python
sub = soup.find_all('td')

sub_th = []

for s in range(4,len(sub),7):
    sub_th.append(sub[s].get_text().replace('\n',''))
```

```python
sub_th
```

```python
lat = soup.find_all('td')

lat_th = []

for xlat in range(5,len(lat),7):
    lat_th.append(lat[xlat].get_text().replace('\n',''))
```

```python
lat_th
```

```python
log = soup.find_all('td')

log_th = []

for ylog in range(6,len(log),7):
    log_th.append(log[ylog].get_text().replace('\n',''))
```

```python
log_th
```

```python
data_set = {'District':district_list,
                           'PostCode':postal,
                           'Thai':desc_th,
                           'Population':pop_th,
                           'Subdistricts':sub_th,
                           'Latitude':lat_th,
                           'Longitude':log_th}

df = pd.DataFrame(data_set)
```
